```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import network_design as nd
import old_network_design as old_nd
import time
from matplotlib.colors import to_rgba
```

```python
O = np.array([[0,2],[2,0]])
I = np.array([[1,0],[1,0],[0,1],[0,1]])
X = np.tile(I,(50,1))
```

```python
maximally_connected = nd.NetAssembly(X,O,X.sum(axis=0))
```

```python
maximally_connected.run(1000,detachment_rate=0.0,connection_method='maximally_connect')
```

```python
maximally_connected.draw(with_networkx=False,node_size=10,color_key=[to_rgba('deepskyblue'),to_rgba('red')])
```
