```python
def write_edgelist_with_header(filename, graph):
    """
    Write an edge list file with a header line containing N and L.

    Parameters:
        filename (str): Name of the output file.
        graph (nx.Graph): A NetworkX graph object.
    """
    num_nodes = graph.number_of_nodes()
    num_edges = graph.number_of_edges()
    
    # Write the edge list using NetworkX
    nx.write_edgelist(graph, filename, data=False)
    
    # Read the file back and add the header line
    with open(filename, 'r') as f:
        edges = f.readlines()
    
    with open(filename, 'w') as f:
        # Write the header line
        f.write(f"{num_nodes} {num_edges}\n")
        # Write the edges
        f.writelines(edges)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
time = np.linspace(0, 10, 1000)  # Time vector
dt = time[1] - time[0]
natural_freq_initial = 0.8  # Initial oscillator frequency (Hz), before perturbation
natural_freq_entrain = 1.0  # Target frequency for entrainment (Hz)
stim_freq_entrain = 1.0  # Stimulation frequency for entrainment (Hz)

natural_freq_sync = 1.0  # Target frequency for synchronization attempt (Hz)
stim_freq_sync = 1.1  # Stimulation frequency for synchronization without full entrainment (Hz)
stim_freq_sync_blue = 1.8
perturb_start_time = 2.0 # Time when perturbation starts
perturb_start_idx = np.searchsorted(time, perturb_start_time)
natural_freq = 1.0 

# Coupling strengths
K_entrain = .5  # Strong coupling for full entrainment
K_sync = 0.0    # Weak coupling for partial synchronization

# Initialize phase arrays
phase_entrain = np.zeros_like(time)
phase_sync = np.zeros_like(time)

# Initial phases (different starting frequencies)
initial_freq_offset = -0.2  # Offset in Hz for initial desynchronization
phase_entrain[0] = 0.0
phase_sync[0] = 0.0

# External stimulation phases
phase_stim_entrain = 2 * np.pi * stim_freq_entrain * time
phase_stim_sync = 2 * np.pi * stim_freq_sync * time

# Functions to create oscillator behavior
for i in range(1, len(time)):
    if i < perturb_start_idx:
        # Before perturbation: free evolution with offset
        phase_entrain[i] = phase_entrain[i-1] + 2 * np.pi * (natural_freq + initial_freq_offset) * dt
        phase_sync[i] = phase_sync[i-1] + 2 * np.pi * (natural_freq + initial_freq_offset) * dt
    else:
        # After perturbation: coupled dynamics
        dtheta_entrain = 2 * np.pi * natural_freq + K_entrain * np.sin(phase_stim_entrain[i-1] - phase_entrain[i-1])
        dtheta_sync = 2 * np.pi * natural_freq + K_sync * np.sin(phase_stim_sync[i-1] - phase_sync[i-1])
        phase_entrain[i] = phase_entrain[i-1] + dtheta_entrain * dt
        phase_sync[i] = phase_sync[i-1] + dtheta_sync * dt

# Create signals
oscillator_entrain = np.sin(phase_entrain)
oscillator_sync = np.sin(phase_sync)
stimulation_entrain = np.sin(2 * np.pi * stim_freq_entrain * time)

stimulation_sync = np.sin(2 * np.pi * stim_freq_sync * time)
stimulation_sync_blue = np.sin(2 * np.pi * stim_freq_sync_blue * time)

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(12, 7), sharex=True)

# Entrainment Plot
axes[0].plot(time, oscillator_entrain, label="Oscillator", color="blue")
axes[0].plot(time[time>2], stimulation_entrain[time>2], label="External Perturbation", color="red", linestyle="dashed")
axes[0].axvline(perturb_start_time, color='gray', linestyle='dotted', label='Perturbation Start')
axes[0].set_title("Entrainment (Phase Locking)")
axes[0].legend()

# Synchronization Plot
axes[1].plot(time[time<2], oscillator_sync[time<2], label="Oscillator", color="blue")
axes[1].plot(time[time>2], stimulation_sync_blue[time>2], color="blue")
axes[1].plot(time[time<8]+2, stimulation_sync[time<8], label="External Perturbation", color="red", linestyle="dashed")
axes[1].axvline(perturb_start_time, color='gray', linestyle='dotted', label='Perturbation Start')
axes[1].set_title("De-Synchronization")
axes[1].legend()

plt.xlabel("Time (s)")
plt.tight_layout()
plt.savefig('summer_fig.png')
plt.show()
```

```python
def likelihood(A, X):
    P = np.ones_like(X)
    for j in range(1, X.shape[1]):
        interaction = A @ X[:, j-1]  # Vectorized computation of the interaction term
        P[:, j] = np.exp(X[:, j] * interaction) / (2 * np.cosh(interaction))
    return np.prod(P)
```

```python
def mcmc(m, X, T, initial=None, burn_in=1000):
    found_graphs = []
    count = {}
    if initial is None:
        g = nx.gnm_random_graph(X.shape[0], m)
        A = nx.adjacency_matrix(g).todense()
    else:
        g = initial
        A = nx.adjacency_matrix(g).todense()

    like0 = likelihood(A, X)
    
    # Burn-in period
    for i in range(burn_in):
        # Create candidate
        new_g = random_rewire(g.copy())
        new_A = nx.adjacency_matrix(new_g).todense()
        like1 = likelihood(new_A, X)
        # Calculate acceptance ratio
        alpha = like1 / like0
        u = np.random.random()
        if u <= alpha:
            g = new_g
            like0 = like1

    # Sampling period
    for i in range(T):
        # Create candidate
        new_g = random_rewire(g.copy())
        new_A = nx.adjacency_matrix(new_g).todense()
        like1 = likelihood(new_A, X)
        # Calculate acceptance ratio
        alpha = like1 / like0
        u = np.random.random()
        if u <= alpha:
            g = new_g
            like0 = like1

        # Record the graph
        graph_tuple = tuple(map(tuple, nx.adjacency_matrix(g).todense()))
        if graph_tuple in count:
            count[graph_tuple] += 1
        else:
            count[graph_tuple] = 1

    # Calculate P(x)
    total_samples = sum(count.values())
    P_x = {graph: freq / total_samples for graph, freq in count.items()}

    return P_x
```
