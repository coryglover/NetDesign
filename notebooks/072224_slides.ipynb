```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import copy
import scipy
import network_design as nd
from matplotlib import cm
```

```python
def color_nodes(X):
    rainbow = cm.get_cmap('gist_rainbow',X.shape[1])
    colors = np.array([rainbow(np.where(X[i]==1)[0]) for i in range(X.shape[0])])
    return colors
```
